# Vector Representation

$\vec{a} = a_x \hat{e}_x + a_y \hat{e}_y$

$a_x, a_y$ = components

$\hat{e}_x, \hat{e}_y$ = unit vectors pointing in $x$ and $y$

Vector Magnitude

$|\vec{a}| = \sqrt{a_x^2+a_y^2}$

Unit Vector

$\hat{e}_a = \frac{\vec{a}}{|\vec{a}|}$

# Vector Addition

$\vec{a} + \vec{b}$

## Component-wise Addition
$\vec{a}+\vec{b} = (a_x + b_x)\hat{e}_x + (a_y + b_y)\hat{e}_y$

```python
import numpy as np
from vectors import add


vector_a = np.array([1, 2])
vector_b = np.array([4, 5])

print(np.array(add(vector_a, vector_b)))
```
